```python
for nom in liste_dataset:
    chemin='D:/m2/reduc/{}.csv'.format(nom)
    import time
    dataset=pd.read_csv(chemin) 
    if(dataset.shape[1]==4):
        X=dataset.iloc[:,1:3]
        label=dataset.iloc[:,3].values
        dim=2
        
        
    else:
        X=dataset.iloc[:,1:4]
        label=dataset.iloc[:,4].values 
        dim=3
        
        
    nombre_classe=len(np.unique(label))
    input_img = Input(shape=(dim,), name="inp_img")

    encoded = Dense(dim, activation='relu')(input_img)

    bottlenek_layer=Dense(2, activation='relu', name="bottleneck_layer")(encoded)

    output = Dense(dim, activation='sigmoid', name="logits")(bottlenek_layer)
    autoencoder = Model(input_img, output )
    autoencoder.compile(optimizer='adam', loss='binary_crossentropy')
    autoencoder.fit(X, X,
                epochs=50,
                batch_size=256,
                shuffle=True)
    Y = autoencoder.predict(X)
    fig = plt.figure(figsize=(30, 18))
    
   


    ax = fig.add_subplot(251, projection='3d')
    if(dataset.shape[1]==4):
        ax.scatter(X.iloc[:,0].values, X.iloc[:,1].values, c=label, cmap=plt.cm.Spectral)
    else:
         ax.scatter(X.iloc[:,0].values, X.iloc[:,1].values, X.iloc[:,2].values, c=label, cmap=plt.cm.Spectral)   
    ax.view_init(4, -72)


    ax = fig.add_subplot(252 )
    plt.scatter(Y[:, 0], Y[:, 1], c=label, cmap=plt.cm.Spectral)
    plt.title("Deep AE ")
    ax.xaxis.set_major_formatter(NullFormatter())
    ax.yaxis.set_major_formatter(NullFormatter())
```
